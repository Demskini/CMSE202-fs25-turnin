## Part 2: Graph Theory and Networks (7 points)

&#9989; **Question 2.1 (9 points)**: We want you to make a very basic model of the personal network you have built this term. Using Networkx, visualize a network that has at least five (including yourself) CMSE202 classmates you have met and worked with this term. There should be an edge connecting you to all of the other classmates. Add at least one edge between two classmates that know each other (besides yourself). If you do not know names, use your group project member names to make this graph. The Day 10 ICA and PCA should help with doing this. I reccomend using dictionaries to make your network. See if you can display the names of this network on your diagram as well. 

````python
import networkx as nx
import matplotlib.pyplot as plt

#stating variable name for nx.graph
G = nx.Graph()

#list initializing classmate names, then making nodes of the content in the list. 
classmates = ["Nick", "Alex", "Case", "Liam", "Kevin", "Aaron"]
G.add_nodes_from(classmates)


'''
Used chagt gpt on 5/1/25 for command to add edges i used in loop and outside of loop.
"Using networkx in python, how do I add edges between names that are either in a list or specific 'objects' in the list?"
'''
#looping through nodes made from list 
for i in classmates:
    if i != "Nick": #if name in classmate list is not nick then add edge between index and nick [I'm nick so i can't be edge to myself]
        G.add_edge("Nick", i)


#adding edges between group members
G.add_edge("Alex", "Liam") 
G.add_edge("Kevin", "Case") 
G.add_edge("Aaron", "Case") 

'''
used chatgpt on 5/1/25 for command below to visualize the graph created: "how do I take a graph I created using networkx, and 
have it displayed us matplotlib?"
'''
pos = nx.spring_layout(G) 
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos, width=2)
plt.show()
````
