### Graph Creation

&#9989; **Question 3.2 (7 points)**: Create a NetworkX graph object from the territory definitions. The territories are the nodes of the graph. Implement a function that builds the graph and adds appropriate attributes for continents. _Hint:_ Look up the documentation of `add_node`, what are the possible inputs? Don't forget to connect each territory to its neighbors. Make sure to add plenty of comments in your code.

```python
import networkx as nx #importing networkx to make a networkx graph

def create_risk_graph(territories):
    """
    Create a NetworkX graph from the Risk territory definitions.
    
    Parameters:
        territories (dict): Dictionary containing territory and connection information
        
    Returns:
        G (nx.Graph): NetworkX graph representing the Risk board
    """
    #initialize empy graph
    G = nx.Graph()
    
    #looping through each continent and connecting territories 
    for continent, territories_dict in territories.items(): #for loop of continentsin sought after list (using items to iterate items in the key)
        for territory, neighbors in territories_dict.items():
            #each territory is created as a node, setting continent to have a unique value to other continents 
            G.add_node(territory, continent=continent)
            
            #looping to add edge to nodes (graph theory)
            for neighbor in neighbors:
                #when looping if the neibhor doesn't have a node, create it!
                if not G.has_node(neighbor):
                    G.add_node(neighbor, continent=continent)
                #adding the edge for neighbors in the terriroties 
                G.add_edge(territory, neighbor)
    
    #G is the initialized empty graph, return an updated graph for the value
    return G

#creating the graph for G
G = create_risk_graph(risk_territories)

'''
below is a print statement to show that there are actually edges and nodes being created
'''
print("Nodes in the graph:")
print(G.nodes(data=True)) #data = True means the the nodes will list there associated values aka the continent.

print("\nEdges in the graph:")
print(G.edges())  
```

### Graph Visualization

&#9989; **Question 3.3 (4 points)**: Create a visualization of the Risk board graph. You can use AI to help you with this. The function below should:

- make a plot of the risk graph.
- color nodes by continent
- allow the user to choose a type of layout. _Hint:_ search networkx documentation for the possible layout. Min 5 layouts should be available.
- print the name of each territory on the respective node
- make a legend for the graph 
- have very descriptive comments 
- increase the size of each node from the default value  
- increase the font size of label from the default value

```python
import matplotlib.pyplot as plt
import networkx as nx
import random

def visualize_risk_graph(G):
    """
    Create a visualization of the Risk board graph with continents colored differently.
    
    Parameters:
        G (nx.Graph): NetworkX graph of the Risk board
    """
    #dictionary to give each continent a color on the colormap
    continent_colors = {
        'North America': 'lightblue',
        'South America': 'green',
        'Europe': 'orange',
        'Africa': 'red',
        'Asia': 'purple',
        'Australia': 'yellow'
    }

    #this assigns the node color based on what continet is there. aka mainland belonging 
    node_colors = [continent_colors[G.nodes[node]['continent']] for node in G.nodes]

    #Print statements to allow userfriendlyness
    print("Choose a layout: ")
    print("1. spring_layout")
    print("2. circular_layout")
    print("3. shell_layout")
    print("4. kamada_kawai_layout")
    print("5. planar_layout")
    #used chat gpt 3/21/25 to help understand what 'input' is and how to make the input and integer.
    layout_choice = int(input("Enter the number of the layout you want to use (1-5): "))
    '''
    below is a string of if statements to connect the input by user to a code that selected the chosen layout choice
    '''
    
    if layout_choice == 1:
        pos = nx.spring_layout(G)
    elif layout_choice == 2:
        pos = nx.circular_layout(G)
    elif layout_choice == 3:
        pos = nx.shell_layout(G)
    elif layout_choice == 4:
        pos = nx.kamada_kawai_layout(G)
    elif layout_choice == 5:
        pos = nx.planar_layout(G)
    else:
        print("Invalid choice, using spring_layout by default.") #if no valid choice given, auto assign spring layout
        pos = nx.spring_layout(G)

    #plotting the graph
    plt.figure(figsize=(12, 12))  # increased figure size and font size just like you hw askes above
    nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1000, font_size=12, font_weight='bold')

    # title for graph with larger font
    plt.title("Risk Board Graph - Territories and Continents", fontsize=16)

    #Legend creation and unique look | Used Chat gpt to help me with the command like marker face color = and looping through continents to be in legend 3/21/25
    continent_legend = list(continent_colors.keys())
    continent_labels = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=continent_colors[continent], markersize=10) for continent in continent_legend]
    plt.legend(continent_labels, continent_legend, title="Continents", loc="upper left", fontsize=12)

    #plot showing
    plt.show()

#function to visualize 
visualize_risk_graph(G)
```
