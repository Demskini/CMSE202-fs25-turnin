````python
class Planet:
    """
    A class representing a planet orbiting a star.

    Attributes:
        mass (float): The mass of the planet.
        orbital_radius (float): The distance of the planet from its star.
        star_mass (float): The mass of the star the planet orbits.
    """
    def __init__(self, mass, orbital_radius, star_mass):
        """
        Initialize a Planet instance with mass, orbital radius, and the mass of its star.
        """
        self.mass = mass
        self.orbital_radius = orbital_radius
        self.star_mass = star_mass
        
    def get_orbital_period(self): #add the missing parameters here
        """
        Calculate and return the orbital period of the planet.
        """
        G = 6.674e-11
        P_seconds = math.sqrt((4 * math.pi**2 * self.orbital_radius**3) / (G * self.star_mass))
        P_years = P_seconds / 3.154e7 #converting large number (secs) to (years)
        return P_years
        pass
````

## Modeling a system of 2 planets around a star

If we have a system of 2 planets orbiting a star, these planets interact with each other gravitationally. This force is proportional to the planets' masses (bigger mass -> stronger force/interaction) and inversely proportional to the distance between the planets (bigger distance -> weaker force/interaction). If this force is substantial, it can move the planets from their initial orbits, even make them collide in extreme cases! To make sense of how this happens, we can start from a simplified model of 2 planets orbiting the same star. In this part you will create and run that model, plot how the radius from the host star changes for each planet.

✅ **Question 2.4 (4 points):** Add docstrings to the methods below. You should include a description of what the method does, a description of parameters and and what the method returns.

````python
def calculate_forces(planet1, planet2):
    ## NOTE: this finds the force planet2 exerts on planet1.
    '''This calculates the forces exerted on planet 1 from planet two.
    It takes the x axis of both planets current locations and y axis of both planets current locations, it uses the information
    of the locations and inteserts it into the formulas to calculate force x and force y.
    Both forces are used in formula to calculate the radial force.
    
    Parameters: (planet1) inserts the first planet you want to have the influences of the other planet on into the function
                (planet2) inserts the second planet into the formula allow you to have planet 2 in the formulas belo.
                
    Return: this returns the radial force acted one planet1 from planet 2.'''
    force_x = planet2.mass * planet1.mass * (planet2.x - planet1.x) / ((planet2.x - planet1.x)**2 + (planet2.y - planet1.y)**2)**(1.5)
    force_y = planet2.mass * planet1.mass *  (planet2.y - planet1.y) / ((planet2.x - planet1.x)**2 + (planet2.y - planet1.y)**2)**(1.5)
    force_radial = force_x * planet1.x/planet1.radius + force_y * planet1.y/planet1.radius
    return force_radial

def take_step(planet, force, delta_time):
    ##NOTE: this function calculates the step that planet takes in delta_t time of the simulation and updates the properties of the planet.
    ##There is no return because the changes are done to the planet directly.
    """ 
    take_step is a function that calculates where the planet will be after the "step" in time. 
    it calculates the acceleration of the planet, the radius change, the velocity change, the new axis of x and y on the planet and the radial velocity.
    It the planets radius is less than 0.001, it sets the planet radius to zero and terminates the planet.
    it also calculates the planets orbital period time. 

    Parameters: (planet) is the planet you want to run the take step function on.
                (force) is where the total radial force is inputed to be in the function. 
                (delta_time) is where you put the chance in time for the simulation to be run in the function.

    Return: this doesn't have an output, it simply updates the values of the class. 
    """
    
    acceleration = force / planet.mass
    radius_change = 0.5 * acceleration * delta_time**2 + planet.velocity_radial * delta_time
    velocity_change = acceleration * delta_time
    planet.radius = planet.radius + radius_change
    planet.x = planet.radius * np.cos(2 * np.pi * delta_time / planet.orbital_period)
    planet.y = planet.radius * np.sin(2 * np.pi * delta_time / planet.orbital_period)
    planet.velocity_radial = planet.velocity_radial + velocity_change
    
    if planet.radius < 0.001:
        planet.radius = 0
        planet.status = "Terminated"
    
    planet.orbital_period = 2 * np.pi * np.sqrt(planet.radius**3/planet.star_mass)
````

```{r, tidy=FALSE, eval=FALSE, highlight=FALSE }
Class: System
    Represents a planetary system with a central star and two orbiting planets.

Attributes:
    - star_mass (float): Set to the mass of the central star.
    - planet1 (Planet): Set to the first planet in the system.
    - planet2 (Planet): Set to the second planet in the system.
    - planet1_pos (list): Initialize as a list of lists with first element [0.0, planet1.radius], tracking planet1's position.
    - planet2_pos (list): Initialize as a list of lists with first element [0.0, planet2.radius], tracking planet2's position.

Methods:
    - __init__:
        Assigns the default values to the attributes.
    - evolve(time):
        Simulate the evolution of the system over a given time.
        - Initialize:
            - t (float): Set to 0.0.
            - delta_time (float): Set to planet1.orbital_period / 5.
        - While t < time:
            - Initialize:
                - total_force_planet1 (float): Set to 0.0.
                - total_force_planet2 (float): Set to 0.0.
            - If both planets are "Active" and their squared distance >= 0.02:
                - Compute total_force_planet1 using calculate_forces(planet1, planet2).
                - Compute total_force_planet2 using calculate_forces(planet2, planet1).
                - Update planet1 position using take_step(planet1, total_force_planet1, delta_time).
                - Update planet2 position using take_step(planet2, total_force_planet2, delta_time).
            - Increment t by delta_time.
            - Append [t, planet1.radius] to planet1_pos.
            - Append [t, planet2.radius] to planet2_pos.
            - If either planet’s status is "Terminated", break the loop.
```

````python
# your code here
import math
import numpy as np

class Planet:

    def __init__(self, mass, radius, star_mass):
        self.mass = mass
        self.radius = radius
        self.star_mass = star_mass
        self.orbital_period = 2 * math.pi * math.sqrt(radius**3 / star_mass)
        self.velocity_radial = 0.0
        self.x = radius
        self.y = 0.0
        self.status = "Active"

        if self.radius < math.sqrt(0.02): #termination of planet for non infinite simulatioin
            self.status = "Terminated" 

class System:

    def __init__(self, star_mass, planet1, planet2):
        
        self.star_mass = star_mass
        self.planet1 = planet1
        self.planet2 = planet2
        self.planet1_pos = [[0.0, planet1.radius]] #to initialize a list of list for position 1 plant pos
        self.planet2_pos = [[0.0, planet2.radius]] #same but for planet 2 pos

    def calculate_forces(self, planet1, planet2): #code taken from above / influenced
       
        force_x = planet2.mass * planet1.mass * (planet2.x - planet1.x) / ((planet2.x - planet1.x)**2 + (planet2.y - planet1.y)**2)**(1.5)
        force_y = planet2.mass * planet1.mass *  (planet2.y - planet1.y) / ((planet2.x - planet1.x)**2 + (planet2.y - planet1.y)**2)**(1.5)
        force_radial = force_x * planet1.x/planet1.radius + force_y * planet1.y/planet1.radius
        return force_radial

    def take_step(self, planet, force, delta_time): #code taken from above / influenced 
       
        acceleration = force / planet.mass
        radius_change = 0.5 * acceleration * delta_time**2 + planet.velocity_radial * delta_time
        velocity_change = acceleration * delta_time
        planet.radius = planet.radius + radius_change
        planet.x = planet.radius * np.cos(2 * np.pi * delta_time / planet.orbital_period)
        planet.y = planet.radius * np.sin(2 * np.pi * delta_time / planet.orbital_period)
        planet.velocity_radial = planet.velocity_radial + velocity_change
        
        if planet.radius < math.sqrt(0.02):
            planet.radius = 0
            planet.status = "Terminated"
        
        planet.orbital_period = 2 * np.pi * np.sqrt(planet.radius**3 / planet.star_mass)

    def evolve(self, time):
       
        t = 0.0 
        delta_time = self.planet1.orbital_period / 5  #change in time to be one fifth of an orbital period for the planet.

        while t < time:
            total_force_planet1 = 0.0
            total_force_planet2 = 0.0

            if self.planet1.status == "Active" and self.planet2.status == "Active": #if both planets active
                distance_squared = (self.planet1.x - self.planet2.x)**2 + (self.planet1.y - self.planet2.y)**2 #calculate distance squared
                if distance_squared >= 0.02: #if distance calculated above is greater than or equal to 0.02
                    total_force_planet1 = self.calculate_forces(self.planet1, self.planet2)
                    total_force_planet2 = self.calculate_forces(self.planet2, self.planet1)
                    self.take_step(self.planet1, total_force_planet1, delta_time)
                    self.take_step(self.planet2, total_force_planet2, delta_time)

            t += delta_time #sets t to be the change in time plus istelf.

            
            self.planet1_pos.append([t, self.planet1.radius]) #updates position for each step for planet 1 
            self.planet2_pos.append([t, self.planet2.radius]) #same but for planet 2 

       
            if self.planet1.status == "Terminated" or self.planet2.status == "Terminated": #if either planet is terminated break it.
                break
````

### Modeling a system of 1 big planet and N small planets around a star

Now let's extend our model for a system of 1 big planet (by big we mean Jupiter size planet) and a bunch of small planets around it, which will be an approximation for the disk.

Here we will provide the generalized code that runs for a set number of planets.


````python

class System:
    def __init__(self, star_mass, planet_list):
        # the primaty planet we care about is the first planet in the list
        self.star_mass = star_mass
        
        self.planet_list = planet_list
        
        self.number_of_planets = len(planet_list)
        
        self.plot_positions = [[0.0, planet_list[0].radius]]
    
    def evolve(self, time):
        t = 0.0
        delta_time = self.planet_list[0].orbital_period/5
        while t < time:
            # delta_t = self.planet_list[0].orbital_period/5 ## this is the step that chooses the resolution of our simulation
            total_force_list = np.zeros(self.number_of_planets)
            for i, planet in enumerate(self.planet_list):
                if planet.status == "Active":
                    total_force_radial = 0.0
                    for other_planet in self.planet_list:
                        if other_planet.status == "Active":
                            if other_planet != planet and (other_planet.x - planet.x)**2 + (other_planet.y - planet.y)**2 >= 0.02:
                                total_force_radial += calculate_forces(planet, other_planet)
                    total_force_list[i] = total_force_radial

            for i, planet in enumerate(self.planet_list):
                if planet.status == "Active":
                    take_step(planet, total_force_list[i], delta_time)

            
            for planet in self.planet_list:
                if planet.radius < 0.02:
                    planet.status == "Terminated"
            if self.planet_list[0].status == "Terminated":
                break
            t += delta_time
            self.plot_positions.append([t, self.planet_list[0].radius])
            
````

Run the simulation for time of 10.

````python
import numpy as np 
'''
Used Chat gpt 3/21/25, It is obvious below that I pasted the results below from chat gpt. I made my own code trying to run it using 
the system that is given to us, but it doesn't work since my code wasn't made to be looped through a list, I had help from chat gpt
asking "how can I transform this code to access the planet in a list" and it spit out slightly different code, I then made changes to my 
code and it was not working, I was getting lots of errors because of my naming system not working the the naming system of the system.

I asked chat gpt to make all the variable names work together and Then I just pasted the results in here because My intial coding was sloppy
and there were many changes that needed to happen.
'''

# Define a function to calculate the gravitational force between two planets
def calculate_forces(planet1, planet2):
    G = 6.67430e-11  # Gravitational constant
    dx = planet2[0] - planet1[0]  # Planet2 - Planet1 (x-coordinate)
    dy = planet2[1] - planet1[1]  # Planet2 - Planet1 (y-coordinate)
    distance = np.sqrt(dx**2 + dy**2)
    force_magnitude = G * planet1[2] * planet2[2] / distance**2  # F = G * (m1 * m2) / r^2
    force_direction = np.array([dx, dy]) / distance  # unit vector
    force = force_magnitude * force_direction
    return force

# Define a function to take a step in the simulation for a given planet
def take_step(planet, force, delta_time):
    ax = force[0] / planet[2]  # acceleration = force / mass
    ay = force[1] / planet[2]
    planet[3] += ax * delta_time  # Update velocity in x-direction
    planet[4] += ay * delta_time  # Update velocity in y-direction
    
    # Update position based on velocity
    planet[0] += planet[3] * delta_time  # Update x position
    planet[1] += planet[4] * delta_time  # Update y position

# Simulation function that runs the planetary system
def run_simulation(star_mass, num_planets, total_time):
    # Initialize planets
    planets = []
    planets.append([1.0, 0.0, 150, 0.0, 0.0])  # Primary planet with mass 150 at radius 1, no initial velocity

    # Create remaining planets
    for i in range(1, num_planets):
        radius = 0.01 + (0.7 - 0.01) * (i / (num_planets - 1))
        planets.append([radius, 0.0, 0.1, 0.0, 0.0])  # Planets with mass 0.1 and evenly spaced radii
    
    # Initialize time variables
    t = 0.0
    delta_time = planets[0][0] / 5  # Time step based on the first planet's radius
    plot_positions = [[t, planets[0][0]]]  # To store the primary planet's radius over time

    # Run the simulation
    while t < total_time:
        forces = np.zeros((num_planets, 2))  # Force components in x and y

        # Calculate forces between all pairs of planets
        for i in range(num_planets):
            for j in range(num_planets):
                if i != j:
                    force = calculate_forces(planets[i], planets[j])
                    forces[i] += force  # Sum forces on planet i

        # Update positions and velocities
        for i in range(num_planets):
            take_step(planets[i], forces[i], delta_time)

        # Check termination condition (if any planet's radius goes below 0.02)
        for i in range(num_planets):
            if planets[i][0] < 0.02:
                planets[i][2] = 0  # Set mass to 0 (effectively terminating the planet)

        if planets[0][2] == 0:  # If primary planet is terminated, break the simulation
            break
        
        t += delta_time
        plot_positions.append([t, planets[0][0]])  # Record the radius of the primary planet

    return plot_positions

# Run the simulation with the given parameters
plot_data = run_simulation(star_mass=400, num_planets=51, total_time=10)


````
